```python
import matplotlib.pyplot as plt
from time import time

# Start time
get_time = time()

# Function to create array
def get_array_of(n):
    array = list(range(n))
    return array

# Generate list
List1 = get_array_of(10)

# End time
end_time = time()

# Calculate total time
Total_time = end_time - get_time

# Print results
print(List1)
print("Execution Time:", Total_time)

# Plot (x = array size, y = execution time)
plt.figure(figsize=(8, 5))
plt.title('BIG O Notation')
plt.plot([len(List1)], [Total_time], '*', color='red')
plt.xlabel("Array Size")
plt.ylabel("Execution Time (seconds)")
plt.show()
```

```python
import matplotlib.pyplot as plt
from time import time

# Function to loop through array
def print_all_elements(arr):
    for item in arr:
        pass

sizes = [100, 500, 1000, 2000]
times = []

for n in sizes:
    arr = list(range(n))
    start = time()
    print_all_elements(arr)
    end = time()
    times.append(end - start)

plt.figure(figsize=(8, 5))
plt.title('Task 2: Linear Time O(n)')
plt.plot(sizes, times, 'o-')
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import time

def print_pairs(arr):
    # Nested loops to simulate O(n^2)
    for i in arr:
        for j in arr:
            _ = i * j  # small dummy operation

# Input sizes
sizes = [500, 1000, 2000, 3000]
times = []

for n in sizes:
    arr = list(range(n))

    start = time.time()
    print_pairs(arr)
    end = time.time()

    times.append(end - start)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(sizes, times, 'o-', color='blue')
plt.title("Quadratic Time O(n^2)")
plt.xlabel("Input Size (n)")
plt.ylabel("Execution Time (seconds)")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from time import time

# Binary search implementation
def binary_search(arr, target):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    return -1

sizes = [1000, 10000, 100000, 200000]
times = []

for n in sizes:
    arr = list(range(n))
    target = n  # search for last element
    start = time()
    binary_search(arr, target)
    end = time()
    elapsed = end - start
    times.append(elapsed)
    print(f"Size: {n}, Time: {elapsed:.8f} sec")

# Plot results
plt.figure(figsize=(8, 5))
plt.title('Task 4: Logarithmic Time O(log n)')
plt.xlabel('Input Size (n)')
plt.ylabel('Execution Time (seconds)')
plt.plot(sizes, times, 'o-', color="green")
plt.show()
```
